Discuss the scenarios where each scaling technique is preferred and why.

- Standard Scaling:

Used When:
   - Data is normally distributed or close to normal
   - Algorithms assume data is centered around 0

Why?:
   - Brings data to a common scale with mean = 0 and std = 1
   - Helps in maintaining the shape of the distribution



- Min Max Sacling:

Used When:
   - You need all features in [0, 1] range
   - Dataset doesn’t contain extreme outliers
   - Used in algorithms sensitive to feature scale or neural networks

Why?:
   - Ensures all features contribute equally during model training
   - Neural networks converge faster with values between 0 and 1

```python
from sklearn.preprocessing import LabelEncoder

label_cols = ['relationship', 'marital_status', 'workclass', 'occupation', 'education', 'native_country']

le = LabelEncoder()
for col in label_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])
```

Discuss the pros and cons of One-Hot Encoding and Label Encoding.

One-Hot Encoding:

- Pros:
  - Doesn’t assume any order or priority among categories.
  - Works well with algorithms that don’t handle categorical data directly.

- Cons:
  - Creates many columns → leads to high dimensionality.
  - Increases memory usage and training time.


Label Encoding:

- Pros:
  - Very simple and memory efficient.
  - Keeps the data as a single column.
  - Useful when categories have ordinal meaning.

- Cons:
  - Implies ordinal relationship even if not true.
  - Can mislead algorithms like linear regression, KNN, or SVM.

Instead of treating capital_gain and capital_loss separately, capital_net gives a clear net value of capital flow, which simplifies learning for the model.

4.Feature Selection:

Use the Isolation Forest algorithm to identify and remove outliers. Discuss how outliers can affect model performance.

```python
from sklearn.ensemble import IsolationForest

# Use numeric data only for outlier detection
numeric_data = df_encoded.select_dtypes(include=['float64', 'int64'])

# Fit Isolation Forest
iso = IsolationForest(contamination=0.01, random_state=42)  # 1% contamination
outlier_pred = iso.fit_predict(numeric_data)

# Add outlier flag to the dataframe
df_encoded['outlier'] = outlier_pred

# Filter out the outliers
df_cleaned = df_encoded[df_encoded['outlier'] == 1].drop(columns=['outlier'])
```

How Outliers Affect Model Performance:

- Outliers can shift the decision boundary.
- Affects statistical assumptions in linear models.
- Leads to poor generalization and higher prediction error.
- Outliers distort cluster centers or class boundaries.
